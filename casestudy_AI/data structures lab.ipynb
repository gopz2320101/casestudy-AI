```python
def prime(num):
    
    if num>1:
        
        for i in range(2,num):
            if (num%i) == 0:
                print("entered number is not prime  number")
                print(i,"times",num//i, "is",num)
                break
            else:
                print(num,"is prime")
                
    else:
        print("number is not prime number")
        
        
num = int(input("enter the number"))
prime(num)
        
```

```python
# extend euclidean algorithm gcd 

def gcdExtended(a, b):  
    # Base Case  
    if a == 0 :   
        return b,0,1
             
   
      gcd,x1,y1 = gcdExtended(b%a, a)  
     
    
    x = y1 - (b//a) * x1  
    y = x1  
     
    return gcd,x,y 
      
  

a, b = 35,15
g, x, y = gcdExtended(a, b)  
print("gcd(", a , "," , b, ") = ", g)   
     
  
```
