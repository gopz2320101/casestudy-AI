```python
import pandas as pd
f2 =pd.read_csv('https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv')
type(f2)
```

```python
f2.head()
```

```python
analyze2014=f2[(f2.Year=='1/1/14 0:00') & (f2.Age !='00 to 17')]

analyze2014.head()
```

```python
analyze2014.count()
```

```python
edusum = analyze2014.groupby('Educational Attainment')['Population Count'].sum()
edusum
```

```python
(edusum/analyze2014['Population Count'].sum())*100
```

```python
import numpy as np
pivot=analyze2014.pivot_table(index=['Personal Income'],columns=['Educational Attainment'],aggfunc=np.sum)
pivot
```

```python
my_cols = set(pivot.columns)

for i in my_cols:
    pivot[i] = (pivot[i] / pivot[i].sum()) * 100

pivot
```

```python
pivot.plot(kind = 'barh')
```
