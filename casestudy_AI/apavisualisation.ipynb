```python
import pandas as pd
my_cities = pd.read_csv("Book1.csv")

import plotly.express as px

fig = px.scatter_mapbox(my_cities, lat="lat", lon="lon", hover_name="City", hover_data=["State"],
color_discrete_sequence=["fuchsia"], zoom=6, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
#4. Shipments of Household Appliances: Line Graphs. The file ApplianceShipments.csv contains the series of quarterly shipments (in millions of dollars) of US household appliances between 1985 and 1989.									(5marks)
#Create a well-formatted time plot of the data using Python.
#Does there appear to be a quarterly pattern? For a closer view of the patterns, zoom in to the range of 3500–5000 on the y-axis.
#Create one chart with four separate lines, one line for each of Q1, Q2, Q3, and Q4. (In Python, this can be achieved by add column for quarter and year. Then group the data frame by quarter and then plot shipment versus year for each quarter as a separate series on a line graph. Zoom in to the range of 3500–5000 on the y-axis. Does there appear to be a difference between quarters?
#Create a line graph of the series at a yearly aggregated level (i.e., the total shipments in each year).
```

```python
df_2 = pd.read_csv("https://raw.githubusercontent.com/Gopikrishnan96/master/main/ApplianceShipments.csv")
df_2.head(10)
```

```python
df_2.drop(df_2.columns[[2, 3]], axis=1, inplace=True)
df_2
```

```python
fig = px.line(df_2, x="Quarter", y="Shipments", title='Appliances shipments by quarter')
fig.show()
```

```python
#trend in Q1 performance, is increased from 1985 to 1988 and then showed a decline from 1988 to 1989.
#trend in Q2 peformance, is increased from 1985 to 1987 ,from 1987 to 1988 it was stagnant.and then showed a increase from 1988 to 1989
#trend in Q3 performance,from 1985 to 1986, there is an increase in shipment,and from 1986 to 1987,there is a dcrease in shipment,and then there is an increase from 1988 to 1989
#trend in Q4 performance,from 1985 to 1986 and then a steep increase in shipment from 1986 to 1987 and then a slight decrease  in 1988 and an increase in 1989
```

```python
df_2[['quarter','year']] = df_2['Quarter'].str.split("-", expand = True) 
df_2
```

```python
fig = px.line(df_2, x="year", y="Shipments",color='quarter')
fig.show()
```

```python
df_3 = df_2[['Shipments','year']].groupby('year').sum()
df_3
```

```python
fig = px.line(df_3)
fig.show()
```
