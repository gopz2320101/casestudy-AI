```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pylab as plt
```

```python
wn = pd.read_csv("https://raw.githubusercontent.com/Gopikrishnan96/master/main/Wine.csv")
```

```python
wn.head()
```

```python
wn.shape
```

```python
wn.describe()
```

```python
#1.drop 'type' attribute from dataset and normalize it 
```

```python
wn.drop(columns = "Type",inplace = True)
```

```python
wn.head()
```

```python
#normalise
from sklearn import preprocessing
wn_normalised= preprocessing.scale(wn)
wn_normalised
```

```python
#2.Carryout PCA and analyse the result as done in the dim reduction lab ex completed in the class
```

```python
pca= PCA()
pca.fit(wn_normalised)
```

```python
pca_sum_df = pd.DataFrame({"SD" : np.sqrt(pca.explained_variance_),
                          "POVariance" : pca.explained_variance_ratio_,
                          "CP":np.cumsum(pca.explained_variance_ratio_)})
```

```python
pca_sum_df = pca_sum_df.transpose()
```

```python
pca_sum_df
```

```python
pca_sum_df.columns = ["PC1", "PC2","PC3","PC4","PC5","PC6","PC7","PC8","PC9","PC10","PC11","PC12","PC13"]
```

```python
pca_Comp_df = pd.DataFrame(pca.components_.transpose(),
                                columns=pca_sum_df.columns,
                                index=wn.columns)
```

```python
pca_Comp_df
```

```python
#we 8 components to account more than 90% variable imformation,first 2 components generally give us 55% of it so large
#in imformation change occurs.
```
